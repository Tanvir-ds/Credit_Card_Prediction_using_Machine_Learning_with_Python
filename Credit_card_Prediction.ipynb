```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the dataset from CSV file to Pandas DataFrame
credit_data = pd.read_csv("credit card.csv")
credit_data
```

```python
# Dropping "ID" column from the dataset
credit_data.drop(columns=['ID'],index=1,inplace=True)
credit_data
```

```python
# Finding duplicate values
credit_data.duplicated().sum()
```

```python
# Printing duplicate rows
credit_data[credit_data.duplicated()]
```

```python
# handling duplicate values
credit_data.drop_duplicates(inplace=True)
credit_data.duplicated().sum()
```

```python
# Finding missing values
credit_data.isnull().sum()
```

```python
# Removing only 2 null values from AGE column
credit_data.dropna(inplace=True)
credit_data.isnull().sum()
```

```python
# Statistical information of the dataset
credit_data.describe()
```

```python
# Changing the target column name 
credit_data.rename(columns={'default.payment.next.month':'transactions'},inplace=True)
credit_data
```

```python
# Separating Ligit and Fraudulent Transactions
ligit = credit_data[credit_data['transactions']==0]
fraudulent = credit_data[credit_data['transactions']==1]
```

```python
print(ligit.shape)
print(fraudulent.shape)
```

```python
# Statistical infromation for ligit transactions dataset
ligit.describe()
```

```python
# Statistical infromation for fraudulent transactions dataset
fraudulent.describe()
```

# Under-Sampling
# Build a dataset containing same distribution of Ligit Transactions and Fraudulent Transactions
# Fraudulent Transactions = 6629

```python
ligit_sample = ligit.sample(6629)
print(ligit_sample.shape)
```

```python
# Concatinating two DataFrame 
modified_credit_data = pd.concat([ligit_sample,fraudulent],axis=0)
modified_credit_data
```

```python
modified_credit_data['transactions'].value_counts()
```

```python
# Statistical infromation for modified_credit_data dataset
modified_credit_data.describe()
```

```python
modified_credit_data.groupby('transactions').mean()
```

```python
# Extracting Independent and Dependent variable
x = modified_credit_data.drop(columns=['transactions'],axis=1)
y = modified_credit_data['transactions']
```

```python
x
```

```python
y
```

```python
# Splitting the dataset into training and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,stratify=y) 
```

```python
print(modified_credit_data.shape,x_train.shape,x_test.shape)
```

# Model Training

```python
# training the XGBClassifier model with trainind dataset
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(x_train,y_train)
```

# Model Evaluation

```python
# Prediction for training dataset
predicted_training_data = model.predict(x_train)

# training dataset accuracy
from sklearn.metrics import accuracy_score
score = accuracy_score(predicted_training_data,y_train)
print("Training Dataset Accuracy:",score)
```

```python
# R Squared Error
from sklearn.metrics import r2_score
error = r2_score(y_train,predicted_training_data)
print("R Squared Error for Training Dataset:",error)
```

```python
# Prediction for test dataset
predicted_test_data = model.predict(x_test)

# test dataset accuracy
from sklearn.metrics import accuracy_score
score = accuracy_score(predicted_test_data,y_test)
print("Test Dataset Accuracy:",score)
```

```python
# R Squared Error
from sklearn.metrics import r2_score
error = r2_score(y_test,predicted_test_data)
print("R Squared Error for Test Dataset:",error)
```

# Creating and Visualizing Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(predicted_test_data,y_test)

sns.heatmap(cm,annot=True,fmt='.1f')
```
